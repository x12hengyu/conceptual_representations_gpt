```python
import pandas as pd
```

```python
df = pd.read_csv("pairwise_response_1_7_davinci-003.csv", header=None)
x, y, n = [], [], []
for _, row in df.iterrows():
    n.append(row[1])
    strings = row[0].split()
    x.append(strings[12])
    y.append(strings[14][:-1])

names = set(x)
```

```python
matrix_dict = {name_x: {name_y: 0 for name_y in names} for name_x in names}

# Populate the nested dictionary with values from x, y, and n
for i in range(len(x)):
    matrix_dict[x[i]][y[i]] = n[i]
    
matrix = pd.DataFrame(matrix_dict)
matrix
```

```python
matrix.to_csv("pairwise_response_1_7_davinci_003_matrix.csv")
```
